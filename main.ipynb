```python
DATA_DIR = "./data/"
RESOURCES_DIR = "./resources/"

IMAGES_DIR = DATA_DIR + "images/"
IMAGES_NOBG_DIR = DATA_DIR + "images_nobg/"
IMAGES_TWO_DIR = DATA_DIR + "images_two/"
IMAGES_ZIPPED_DIR = DATA_DIR + "images_zipped/"
IMAGES_ZZIPPED_DIR = DATA_DIR + "images_zzipped/"
IMAGES_DETAIL_DIR = DATA_DIR + "images_detail/"
IMAGES_ZZZIPPED_DIR = DATA_DIR + "images_zzzipped/"
IMAGES_MIN_DIR = DATA_DIR + "images_min/"

VIDEO_PATH = RESOURCES_DIR + "kemu3.mp4"

NPY_PATH = DATA_DIR + "data.npy"
```

```python
import cv2

video = cv2.VideoCapture(VIDEO_PATH)

index = 0
if video.isOpened():
    rval, frame = video.read()
else:
    rval = False
    print("Error opening video stream or file")

while rval:
    rval, frame = video.read()
    if frame is None:
        break
    else:
        cv2.imwrite(IMAGES_DIR + str(index) + ".jpg", frame)
    index += 1
    print("已成功分割出", str(index) + "张图片")
```

```python
import os
from rembg import remove

images = os.listdir(IMAGES_DIR)
images.sort(key=lambda x: int(x.split(".")[0]))

length = len(images)

index = 0
for image in images:
    full_path = os.path.join(IMAGES_DIR, image)
    save_path = os.path.join(IMAGES_NOBG_DIR, image)
    with open(full_path, "rb") as i:
        with open(save_path, "wb") as o:
            __input = i.read()
            output = remove(__input)
            o.write(output)
            index += 1
            print("已成功处理", str(index / length))
```

```python
import cv2

images = os.listdir(IMAGES_NOBG_DIR)
images.sort(key=lambda x: int(x.split(".")[0]))

length = len(images)

index = 0
for image in images:
    full_path = os.path.join(IMAGES_NOBG_DIR, image)
    save_path = os.path.join(IMAGES_TWO_DIR, image)
    __image = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)
    row, col = __image.shape
    for r in range (row - 1):
        for c in range (col - 1):
            if __image[r, c] != 0:
                __image[r, c] = 255
    cv2.imwrite(save_path, __image)
    index += 1
    print("已成功处理", str(index / length))
```

```python
import cv2
import math
import numpy as np

images = os.listdir(IMAGES_TWO_DIR)
images.sort(key=lambda x: int(x.split(".")[0])),

length = len(images)

index = 0
for image in images:
    full_path = os.path.join(IMAGES_TWO_DIR, image)
    save_path = os.path.join(IMAGES_ZIPPED_DIR, image)
    __image = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)
    row, col = __image.shape
    tmp = np.zeros((math.floor(row / 2), math.floor(col / 2)), dtype=np.uint8)
    r = 0
    while r < row and r + 1 < row:
        c = 0
        while c < col and c + 1 < col:
            ave = math.floor(sum([__image[r, c], __image[r + 1, c], __image[r + 1, c + 1], __image[r, c + 1]]) / 4)
            tmp[int(r / 2), int(c / 2)] = 255 if ave > 127 else 0
            c += 2
        r += 2
    cv2.imwrite(save_path, tmp)
    index += 1
    print("已成功处理", str(index / length))
```

```python
import numpy as np

data = np.load(NPY_PATH)
pictures = []

length, row, col = data.shape
for l in range(length):
    tmp = ""
    for r in range(row):
        for c in range(col):
            if data[l, r, c] == 0:
                tmp += "0 "
            else:
                tmp += "1 "
        tmp += "\n"
    pictures.append(tmp)

print(pictures[0])
```
